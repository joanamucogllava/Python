Once a Corpora is loaded/imported. NLTK tool provides a lot of functions to extract information from a corpora. For more details see `Table 2-3`

```python
from IPython.display import Image
Image(filename='Chapter2-NLTKCorpaFunctions.jpg')
```

```python
#this gives the raw content of the file austen-emma.txt in gutenberg
raw = gutenberg.raw('austen-emma.txt') 
print(raw[0:10])
#his gives the words of the file austen-emma.txt in gutenberg
words = gutenberg.words('austen-emma.txt') 
print(words[0:10])
#his gives the sentences in the file austen-emma.txt in gutenberg
sents= gutenberg.sents('austen-emma.txt') 
print(sents[1:10])

#number of words in `austen-emma.txt`
num_words = len(gutenberg.words('austen-emma.txt') )
num_chars = len(gutenberg.raw('austen-emma.txt') )
num_sents = len(gutenberg.sents('austen-emma.txt') )
vocab = set(w.lower() for w in words)
num_vocab = len(vocab)

print("average word length")
print(num_chars/num_words)
print("average sentence length")
print(num_words/num_sents)
print("average number of times each vocabulary item appears in the text: lexical diversity")
print(num_words/num_vocab)
```

You can also use a `for-loop` to extract similar information for all the text/files/books in the corpus.

```python
for field in gutenberg.fileids():
    num_words = len(gutenberg.words(field) )
    num_chars = len(gutenberg.raw(field) )
    num_sents = len(gutenberg.sents(field) )
    vocab = set(w.lower() for w in words)
    num_vocab = len(vocab)
    print(int(num_chars/num_words), int(num_words/num_sents), int(num_words/num_vocab), field)
```

#### Wordlist Corpora


NLTK includes some corpora that are nothing more than wordlists. The Words Corpus
is the /usr/dict/words file from Unix, used by some spellcheckers. We can use it to find
unusual or misspelled words in a text corpus,

```python
nltk.download('words')
```

```python
# a function that computes the vocabulary of a text, removes all items that occur in an existing wordlist, returning just the uncommon/mispelled words
from nltk.corpus import words
def unusual_words(text):
 text_vocab = set(w.lower() for w in text if w.isalpha()) # isalpha simply checks if the word consists of letters only
 english_vocab = set(w.lower() for w in nltk.corpus.words.words())
 unusual = text_vocab.difference(english_vocab)
 return sorted(unusual)
```

```python
unusual_words(gutenberg.words('austen-emma.txt'))
```

#### Stopwords Corpora

There is also a corpus of stopwords, that is, high-frequency words such as `the`, `to`, `and`
`also` that we sometimes want to filter out of a document before further processing.
Stopwords usually have little lexical content, and their presence in a text fails to distinguish it from other texts.

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

Lets define a function to compute what fraction of words in a text are not in the stopwords list:

```python
def content_fraction(text):
   stopwords = nltk.corpus.stopwords.words('english')
   content = [w for w in text if w.lower() not in stopwords]
   return len(content) / len(text)
```

```python
content_fraction(nltk.corpus.gutenberg.words())
```

Remove all stopwords from a `text` and get freqeuncy distribution

```python
stopwords = nltk.corpus.stopwords.words('english')
adventure_text = brown.words(categories='adventure')
text = [word for word in adventure_text if word.lower() not in stopwords and len(word) > 4]
fdist_ad = nltk.FreqDist(text)
fdist_ad.plot(20)
```

#### Simple example of solving puzzle with wordlist

```python
from IPython.display import Image
Image(filename='Chapter2-puzzle.jpg')
```

```python
puzzle_letters = nltk.FreqDist('egivrvonl')
obligatory = 'r'
wordlist = nltk.corpus.words.words()

'''
Three conditions:
length of word >= 4
word contains 'r'
FreqDist comparison permits us to check that the frequency of each letter in the candidate word 
is less than or equal to the frequency of the corresponding letter in the puzzle.
'''

[w for w in wordlist if len(w) >= 4 and obligatory in w and nltk.FreqDist(w) <= puzzle_letters]
```
